```python
def rating(rating):
    if rating>=7.0:
        return 'Exelent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
print('Single Genre of Movies :')
two_d=[]
for value in df['Genre']:
    two_d.append(value.split(','))

one_d=[]
for item in two_d:
    for unque in item:
        one_d.append(unque)

uniq_ls=[]
for item in one_d:
    if item not in uniq_ls:
        uniq_ls.append(item)

print('Length of unique Genra :',len(uniq_ls))
uniq_ls
```
