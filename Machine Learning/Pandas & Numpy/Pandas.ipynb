```python
d={
    0:[1,2,3],
    1:[4,5,6],
    2:[7,8,9]
}
df_new = pd.DataFrame(d)
```

```python

print('Row wise sort : \n')
print(df_new.sort_index(axis=0))

print('\nColoumn wise sort : \n')
print(df_new.sort_index(axis=1))

print('\nRow wise desending sort : \n')
print(df_new.sort_index(axis=0,ascending=False))

print('\nColoumn wise desending sort : \n')
print(df_new.sort_index(axis=1,ascending=False))

print('\nRow wise asending sort : \n')
print(df_new.sort_index(axis=0,ascending=True))

print('\nColoumn wise asending sort : \n')
print(df_new.sort_index(axis=1,ascending=True))
```
